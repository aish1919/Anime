```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Anime.csv")
df.head()
```

```python
df.info()

#With the help of info() function we get the information regarding our DataFrame.
#i.e Total number of columns are 17, titales of the columns, data types of the columns, non-null value count, range index i.e total entries of the column here it is 18495.`
#It is a useful function for the further operations on our datasets i.e Visualization, Finding missing values, Duplicate values etc. 
```

```python
# Using isnull function we get the missing values i.e null values.
#In my datasets following columns have null values which we are going to replaced or dropped using functions.
df.isnull().sum()
```

```python
#here, more than 40% data is missing in these columns so using drop function im dropping these columns. 

df.drop(["Release_season", "Japanese_name","End_year", "Content_Warning", "Related_Mange","Related_anime"],axis=1, inplace=True)
```

```python
#To Check wheather the above mentioned columns got deleted or not.
df.isnull().sum()
```

```python
#using head function we get the first row of our Datasets
df.head()
```

```python
#Here im checking the skewness with the help of distplot.
sns.distplot(df["Episodes"])
```

```python
#Right skewness present in the data that is the reason why im calculating the median to fill the null values.

df["Episodes"].median()
```

```python
#with the help of fillna function i am going to fill the null values with the median i.e 12

df["Episodes"].fillna(df["Episodes"].median(), inplace=True)
```

```python
df.isna().sum()
```

```python
#Checking the counts of unique value with the hepl of value_counts function.
df.Studio.value_counts()
```

```python
#here i have used ffill method i.e forword fill method to fill null values.
df.Studio.fillna(method="ffill", inplace=True)  
```

```python
df.isna().sum()
```

```python
#as per above null value table i can see that there are only 3% data in the columns are missing so im dropping the null values from it.
#dropping the null values using dropna.

df.dropna(inplace=True)
```
